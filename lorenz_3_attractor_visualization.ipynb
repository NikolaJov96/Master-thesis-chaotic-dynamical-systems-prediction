```python
#%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
#plt.rcParams["figure.figsize"] = (10,10)
```

# Lorenz 3

```python
# Integration time step
dt = 0.005
  
# Lorenz ODE parameters

sigma = 10.0
r     = 28.0
b     = 8.0/3.0
  
# Initial conditions
x = -10.0
y =  -7.0
z =  35.0
  
# Store the trajectory
TrajX = []
TrajY = []
TrajZ = []
Time  = []

# Integrate the Lorenz ODEs
#
for t in np.arange(0.,50.,dt):
    dxdt = sigma*(y - x)
    dydt = r * x - y - x * z
    dzdt = x * y - b * z
    
    x = x + dxdt * dt
    y = y + dydt * dt
    z = z + dzdt * dt

    TrajX.append(x)
    TrajY.append(y)
    TrajZ.append(z)
    Time.append(t)

TrajX, TrajY, trajZ, Time = np.array(TrajX), np.array(TrajY), np.array(TrajZ), np.array(Time)
    
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.plot3D(TrajX, TrajY, TrajZ, 'blue')
```
